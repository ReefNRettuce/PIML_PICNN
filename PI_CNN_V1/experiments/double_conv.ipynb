```python
# all I'm doing is try to do double convolution right now to understand the UNET 
# architecture. 

import torch
import torch.nn as nn

def double_convolutions(in_channels, out_channels):
    """
    Double convolution function block. This takes in an input, weight, and produces 
    and produces an output. 

    To make an convolution we call the convolution function conv2d. Then we use the inputs
    and weights as the components to convolve together. We're copying the origianl 
    architecture but we're going to add a pad to the sides. 

    The convolution essentially takes in two input tensors and averages them together (similiar
    to a moving average filter)

    tensor_1 = [1,2,3,4]
    tensor_2 = [1,2,3,4]

    stride is how far we move each time. and padding is how many zeroes we add to the ends.
    
    """

    convolutional_operation = nn.Sequential(
        nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=0),
        nn.BatchNorm2d(out_channels),
        nn.ReLU(inplace=True),
        nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=0),
        nn.BatchNorm2d(out_channels),
        nn.ReLU(inplace=True)
    )

    return convolutional_operation
```
